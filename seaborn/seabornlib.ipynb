Seaborn:
    -Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built-in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for the plot
sns.set_theme()
```

```python
sns.relplot(x='total_bill', y='tip',col='time',hue='smoker', style='smoker',size='size', data=tips)
plt.show()
```

```python
#load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

ScatterPlot

```python
sns.scatterplot(x='sepal_length', y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width',hue='species',data=iris)
```

CountPlot

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
sns.countplot(x='class',hue='class', data=titanic)
```

```python
sns.countplot(x='survived',hue='survived', data=titanic)
```

Bar Chart

```python
sns.barplot(x='sex', y='survived',hue='class',data=titanic)
```

Distribution Plot

```python
sns.displot(titanic['survived'])
```

Correlation

HeatMap

```python
correlation = titanic.corr(numeric_only=True)
```

```python
#constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True,annot_kws={'size':8}, cmap='Blues')
```

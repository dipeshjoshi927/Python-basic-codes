```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

Plotting the data

```python
#sine wave 
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
```

```python
#adding title, x-axis & y-axis lables
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

we can also use plot(x,y,'g.')

```python
#we can use different colors and symbols for plotting
plt.plot(x,y,'r*')
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x),'g-')
plt.plot(x, np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Spanish','German','French','Latin']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('languages')
plt.ylabel('Number of People')
plt.title('Random Bar Graph')
plt.show()
```

Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Spanish','German','French','Latin']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.2f%%')
plt.show()
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)
z= np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3d-Plotting

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(200)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Reds')
plt.show()
```
